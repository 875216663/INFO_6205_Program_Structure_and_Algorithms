# Traveling Salesman Problem (TSP)

The Traveling Salesman Problem (TSP) is a well-known optimization problem where the goal is to find the shortest possible route that visits a given set of cities and returns to the starting city. Approximation algorithms can provide solutions that are close to optimal. One such approach is the Nearest Neighbor Algorithm.

---

Here's an explanation along with a Python code example.

<br>

# Nearest Neighbor Algorithm for TSP
**Problem**: Given a set of cities and the distances between them, find the shortest route that visits each city exactly once and returns to the starting city.

**Algorithm**: The Nearest Neighbor Algorithm starts from a specified city and repeatedly selects the nearest unvisited city as the next destination until all cities are visited. The route forms a cycle that may not be optimal but provides a quick approximation.

**Illustration**:
Consider the following example with four cities (A, B, C, and D) and the distances between them:

1. Step 1: Start from a specific city, say A.<br>
Current city: A<br>
Visited cities: A

2. Step 2: Find the nearest unvisited city from the current city A.<br>
Nearest unvisited city: C<br>
Distance: 15

3. Step 3: Move to city C.<br>
Current city: C<br>
Visited cities: A, C

4. Step 4: Find the nearest unvisited city from city C.<br>
Nearest unvisited city: B<br>
Distance: 20

5. Step 5: Move to city B.<br>
Current city: B<br>
Visited cities: A, C, B

6. Step 6: Find the nearest unvisited city from city B.<br>
Nearest unvisited city: D<br>
Distance: 17

7. Step 7: Move to city D.<br>
Current city: D<br>
Visited cities: A, C, B, D

8. Step 8: Return to the starting city A to complete the cycle.<br>
Current city: A
Visited cities: A, C, B, D, A<br>

**The approximate TSP route is A -> C -> B -> D -> A.**

```python
def nearest_neighbor_tsp(graph, start):
    n = len(graph)
    visited = [False] * n
    route = [start]
    total_distance = 0

    current_city = start
    visited[start] = True

    for _ in range(n - 1):
        nearest_city = None
        min_distance = math.inf

        for neighbor in range(n):
            if not visited[neighbor] and graph[current_city][neighbor] < min_distance:
                nearest_city = neighbor
                min_distance = graph[current_city][neighbor]

        route.append(nearest_city)
        total_distance += min_distance
        current_city = nearest_city
        visited[current_city] = True

    # Return to the starting city to complete the cycle
    route.append(start)
    total_distance += graph[current_city][start]

    return route, total_distance

# Example graph: Distance between cities
graph = [[0, 29, 15, 21],
         [29, 0, 20, 17],
         [15, 20, 0, 20],
         [21, 17, 20, 0]]

start_city = 0
route, total_distance = nearest_neighbor_tsp(graph, start_city)
print("Approximate TSP Route:", route)
print("Total Distance:", total_distance)
```

In the code example, the algorithm starts from city A and proceeds to form a route that visits each city once and returns to A. The total distance of the route is calculated based on the distances between the cities in the given graph.

Keep in mind that the Nearest Neighbor Algorithm is a heuristic and may not always yield the optimal solution. However, it's a quick and simple approach to get an approximate solution for the Traveling Salesman Problem.
